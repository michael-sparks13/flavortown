```python
import requests

import time

from bs4 import BeautifulSoup
```

```python
headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
    }
```

```python
base_url = 'https://www.flavortownusa.com/locations?page='
```

```python
restaurants = {}

for page in range(1,47):
    url = base_url + str(page)
    
    req = requests.get(url, headers)
    
    soup = BeautifulSoup(req.content, 'html.parser')
    
    rests = soup.find_all('div', class_='listing-item-inner')
    
    for rest in rests:
        add = rest.find('span').get_text()
        name = rest.find('h3').get_text()
        
        restaurants[name] = add
    #wait for 1 second
    time.sleep(1)
    


print(restaurants)
```

```python
visit_dates = {}

for page in range(1,47):
    url = base_url + str(page)
    
    req = requests.get(url, headers)
    
    soup = BeautifulSoup(req.content, 'html.parser')
    
    rests = soup.find_all('div', class_='listing-item-inner')
    
    for rest in rests:
        visit_date = rest.select_one('li:nth-child(1)')
        name = rest.find('h3').get_text()
        
        visit_dates[name] = visit_date
    #wait for 1 second
    time.sleep(1)
    


print(visit_dates)
```
